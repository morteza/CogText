```python
import random
random.seed(42)

import matplotlib.pyplot as plt

# a random graph
g = ig.Graph.Erdos_Renyi(10,.02)

# detect communities
comms = g.community_multilevel()

# plot and save
g_plot = ig.plot(comms, mark_groups = True, inline=True)

g_plot.save('outputs/multilevel_community_in_graphs.png')

print(comms)
g_plot
```
