```python
%reload_ext autoreload
%autoreload 3
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
n_rows = 11
n_cols = 11

xy = np.mgrid[0:10:11j, 0:10:11j].reshape(2,-1).T

xy_df = pd.DataFrame(xy, columns=['x','y'])

bg_circle_styles = {
    's': 200,
    'color': 'lightgrey',
}

bg_line_styles = {
    'lw': 3,
    'color': 'lightgrey'
}


fig, ax = plt.subplots(figsize=(15, 10))

ax.scatter(xy_df['x'], xy_df['y'], **bg_circle_styles,zorder=1)

for xi in range(11):
    ax.plot([xi,xi], [0,10], **bg_line_styles,zorder=2)

ax.scatter([6,6,6,4,4], [5,4,2,1,8], color='black', s=200,zorder=3)

ax.plot([6,6], [2,5], lw=3, color='black',zorder=4)
ax.plot([4,4], [1,8], lw=3, color='black',zorder=4)

# alternating row shading (XXX: use add_patch(Rectangle)?)
for i in range(0, 10+1, 2):
    rect = plt.Rectangle((-.5, i-.45), 11, .9, facecolor='#eee', lw=0, zorder=0)
    ax.add_patch(rect)

ax.xaxis.tick_top()
ax.invert_yaxis()

ax.grid(False)
ax.set_frame_on(False)

ax.tick_params(axis='both', which='both',length=0)

plt.xticks(range(11))
plt.yticks(range(11))

[t.set_y(.95) for t in ax.xaxis.get_ticklabels()]

[t.set_x(0.05) for t in ax.yaxis.get_ticklabels()]


ax.set_xticklabels([f'construct{i}' for i in range(11)])
ax.set_yticklabels([f'task{i}' for i in range(11)])

plt.show()
```
