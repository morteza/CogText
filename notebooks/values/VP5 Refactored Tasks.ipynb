This notebooks plots the diagrams for the [5th value proposition](https://www.notion.so/mortynia/VP5-Refactored-task-space-78509192c47b47b29d20ea7d272a6060), refactored task space.

```python
# figure 1: number of tasks per publication

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from python.cogtext.datasets.pubmed import PubMedDataLoader

PUBMED = PubMedDataLoader(preprocessed=False).load()
```

```python
# nunmber of tasks per study

task_per_pub_counts = PUBMED.query('category.str.contains("Task")').groupby('pmid')[['category','label']].count()
task_per_pub_counts['label'] = task_per_pub_counts['label'].astype('category')

# PUBMED.query('category.str.contains("Task")')['pmid'].nunique()

import seaborn as sns
sns.set('paper', 'whitegrid')
sns.displot(task_per_pub_counts['label'],
            color='gray',
            aspect=1.8).set(
  xlabel='# of tasks per publication',
  ylabel='# of publications',
  ylim=(-1000,35000),
  title='From a corpus of 39637 publications that mentioned at least one cognitive task')

sns.lineplot(data=task_per_pub_counts.value_counts().to_frame(), x='label', y=0, lw=2, color='black')

plt.show()
```
